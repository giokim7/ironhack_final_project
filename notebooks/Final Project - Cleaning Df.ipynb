```python
import pandas as pd
```

Import file

```python
tsv_file_path = 'prc_hpi_a.tsv'
df = pd.read_csv(tsv_file_path, sep='\t')
print(df)
```

Sort columns

```python
columns_to_sort = df.columns[1:]

# Sort columns based on their names
df = df[[df.columns[0]] + sorted(columns_to_sort)]
```

Keep rows referring to Total Housing Prices with 2010 as 100 index

```python
df = df[df['Country'].str.startswith('TOTAL,I10_A_AVG,')]
```

Remove TOTAL,I10_A_AVG, from every row

```python
df['Country'] = df['Country'].str.replace('TOTAL,I10_A_AVG,', '')
```

Replace abbreviation with full Country name

```python
country_mapping = {
    'AT': 'Austria',
    'BE': 'Belgium',
    'BG': 'Bulgaria',
    'HR': 'Croatia',
    'CY': 'Cyprus',
    'CZ': 'Czech Republic',
    'DK': 'Denmark',
    'EE': 'Estonia',
    'FI': 'Finland',
    'FR': 'France',
    'DE': 'Germany',
    'GR': 'Greece',
    'HU': 'Hungary',
    'IE': 'Ireland',
    'IT': 'Italy',
    'LV': 'Latvia',
    'LT': 'Lithuania',
    'LU': 'Luxembourg',
    'MT': 'Malta',
    'NL': 'Netherlands',
    'PL': 'Poland',
    'PT': 'Portugal',
    'RO': 'Romania',
    'SK': 'Slovakia',
    'SI': 'Slovenia',
    'ES': 'Spain',
    'IS': 'Iceland',
    'NO': 'Norway',
    'TR': 'Turkey',
    'UK': 'Ukraine',
    'SE': 'Sweden'
}    

df['Country'] = df['Country'].replace(country_mapping)
```

Remove letters and empty spaces

```python
def remove_letters(s):
    return ''.join(filter(lambda x: not x.isalpha(), s))

df.columns = df.columns.str.strip()

# Replace non-numeric values with NaN
df = df.apply(pd.to_numeric, errors='coerce')

# apply the function to all cells except the first row and first column
df.iloc[1:, 1:] = df.iloc[1:, 1:].applymap(remove_letters)
```

Save in new file

```python
file_path = 'housing_prices_EU.csv'  
df.to_csv(file_path, index=False)  
```
